```python
# Drop the first name and surname columns since they don't add anything to the analysis and there's already a user_id column
cust2 = cust.drop(columns=["First Name", "Surnam"])
```

```python
# Let's also make the column headers more consistent
```

```python
# Rename Gender as gender
cust2.rename(columns = {'Gender' : 'gender'}, inplace = True)
```

```python
# Rename STATE as state 
cust2.rename(columns = {'STATE' : 'state'}, inplace = True)
```

```python
# Rename Age as age
cust2.rename(columns = {'Age' : 'age'}, inplace = True)
```

```python
# Rename n_dependants as number_of_dependants
cust2.rename(columns = {'n_dependants' : 'number_of_dependants'}, inplace = True)
```

```python
# Rename fam_status as marital_status
cust2.rename(columns = {'fam_status' : 'marital_status'}, inplace = True)
```
